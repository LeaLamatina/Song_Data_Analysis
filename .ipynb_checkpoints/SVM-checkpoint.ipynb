```python
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import numpy as np
import os
from sklearn import preprocessing
```

```python
df = pd.read_csv(os.path.join("Resources", "Spotify_Audio_Features.csv"))
df.head()
```

```python
#Encode track id into quantitative
le = preprocessing.LabelEncoder()
df['track_id'] = le.fit_transform(df['track_id'].astype('str'))
```

```python
#Encode track_name into quantitative
le = preprocessing.LabelEncoder()
df['track_name'] = le.fit_transform(df['track_name'].astype('str'))
```

```python
#Encode artist_name into quantitative
le = preprocessing.LabelEncoder()
df['artist_name'] = le.fit_transform(df['artist_name'].astype('str'))
```

```python
target = df["popularity"]
target_names = ["negative", "positive"]
```

```python
data = df.drop("popularity", axis=1)
feature_names = data.columns
data.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=42)
```

```python
drop_list = ['artist_name', 'track_name']
train = df.drop(drop_list, axis=1)
```

```python
# Support vector machine linear classifier
from sklearn.svm import SVC 
import timit

start = timit.default_timer()

model = SVC(kernel='linear')
model.fit(X_train, y_train)

stop = timit.default_timer()

print ('Time to run this code (in seconds):', stop - start)
```
